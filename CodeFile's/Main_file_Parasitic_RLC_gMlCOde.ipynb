```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Creating the dataset
data = {
    'Adder': ['RCA', 'CSLA', 'CSA', 'CBA', 'CIA', 'CSkA', 'KSA', 'RCA', 'CSLA', 'CSA', 'CBA', 'CIA', 'CSkA', 'KSA', 'RCA', 'CSLA', 'CSA', 'CBA', 'CIA'],
    'Node': ['AO', 'AO', 'AO', 'AO', 'AO', 'AO', 'AO', 'BO', 'BO', 'BO', 'BO', 'BO', 'BO', 'BO', 'A 1', 'A 1', 'A 1', 'A 1', 'A 1'],
    'C,fF': [7.02, 7.08, 6.58, 7.88, 5.58, 8.89, 4.89, 6.82, 6.88, 6.39, 7.5, 5.39, 9.04, 4.7, 7.67, 6.37, 5.87, 8.51, 6.23],
    'R,ohm': [864, 865, 862, 1386, 855, 1032, 698, 504, 504, 501, 664, 45, 1032, 337, 868, 860, 857, 1390, 859],
    'L,um': [140, 142, 130, 158, 104, 183, 99, 137, 139, 127, 152, 101, 186, 96, 156, 125, 112, 174, 120],
    'PD,uW': [32.16, 73.605, 95.889, 68.83, 42.794, 58.351, 63.537, 32.16, 73.605, 95.889, 68.83, 42.794, 58.351, 63.537, 32.16, 73.605, 95.889, 68.83, 42.794]
}

# Creating a DataFrame from the dataset
df = pd.DataFrame(data)

# Performing K-means clustering with 11 clusters
X = df[['C,fF', 'R,ohm', 'L,um', 'PD,uW']]
kmeans = KMeans(n_clusters=11)
df['Cluster'] = kmeans.fit_predict(X)

# Generating 'K C R L PD CL' dataframe
clustered_data = df.groupby('Cluster').agg({'C,fF': 'mean', 'R,ohm': 'mean', 'L,um': 'mean', 'PD,uW': 'mean', 'Adder': 'count'}).reset_index()
clustered_data.columns = ['K', 'C', 'R', 'L', 'PD', 'CL']
print("K C R L PD CL")
print(clustered_data.to_string(index=False))

# Finding the range of 'C,fF', 'R,ohm', 'L,um', and 'PD,uW' observed from K-means algorithm
range_data = df.groupby('Cluster').agg({'C,fF': ['min', 'max'], 'R,ohm': ['min', 'max'], 'L,um': ['min', 'max'], 'PD,uW': ['min', 'max']})
range_data.columns = ['Min C(fF)', 'Max C(fF)', 'Min R(ohm)', 'Max R(ohm)', 'Min L(um)', 'Max L(um)', 'Min PD(uW)', 'Max PD(uW)']
print("\nRANGE OF C, R, L, PD OBSERVED FROM K-MEANS ALGORITHM (K IS CLUSTER VALUE)")
print(range_data.to_string())

plt.figure(figsize=(8, 6))
plt.scatter(df['R,ohm'], df['PD,uW'], c=df['Cluster'], cmap='viridis', edgecolor='k')
plt.title('K-means Clustering based on Resistance Range')
plt.xlabel('R,ohm')
plt.ylabel('PD,uW')
plt.colorbar(label='Cluster')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression

# Provided dataset (considering a portion of the given data)
data = {
    'Node': ['RCA A0', 'CSLA A0', 'CSA A0', 'CBA A0', 'CIA A0', 'CSkA A0', 'KSA A0',
             'RCA B0', 'CSLA B0', 'CSA B0', 'CBA B0', 'CIA B0', 'CSkA B0', 'KSA B0',
             'RCA A1', 'CSLA A1', 'CSA A1', 'CBA A1', 'CIA A1'],
    'C': [7.02, 7.08, 6.58, 7.88, 5.58, 8.89, 4.89,
          6.82, 6.88, 6.39, 7.5, 5.39, 9.04, 4.7,
          7.67, 6.37, 5.87, 8.51, 6.23],
    'R': [864, 865, 862, 1386, 855, 1032, 698,
          504, 504, 501, 664, 45, 1032, 337,
          868, 860, 857, 1390, 859]
}

# Creating a DataFrame from the provided dataset
df = pd.DataFrame(data)

# Extracting independent (X) and dependent (y) variables
X = df['R'].values.reshape(-1, 1)  # Independent variable (R)
y = df['C'].values  # Dependent variable (C)

# Linear regression model fitting
model = LinearRegression()
model.fit(X, y)

# Predicting capacitance using the trained model
y_pred = model.predict(X)

# Plotting the actual vs predicted capacitance using linear regression
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='Actual C')  # Actual capacitance
plt.plot(X, y_pred, color='red', label='Predicted C')  # Predicted capacitance
plt.xlabel('Resistance (R)')
plt.ylabel('Capacitance (C)')
plt.title('Linear Regression for Capacitance Estimation')
plt.legend()
plt.grid(True)
plt.show()
```
